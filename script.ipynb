```python
import pandas as pd
```

```python
df = pd.DataFrame({
        "RollNo":[1,2,3,4,5],
        "Name": ["Gil","Rakshit","Abhi","Ram","suresh"],
        "Gender": ["male", "male", "male","male","male"],
        "Mark1":[95,100,97,80,75],
        "Mark2":[85,90,76,90,85],
        "Mark3":[65,80,89,60,95]
})
```

```python
df
```

```python
df["TotalMarks"]=df["Mark1"]+df["Mark2"]+df["Mark3"]
```
